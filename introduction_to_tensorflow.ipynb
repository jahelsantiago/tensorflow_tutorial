**Detener los epoch cuando se llegue a cierto valor x** 

cuando se alcance el valor deseado detener el training, se hace mediante callbacks

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs = {}):
    if(logs["loss"] < 0.4):
      print("\n accuary del 60% cosneguida")
      self.model.stop_training = True #detiene el training

callbacks = myCallback()
mnist = tf.keras.datasets.fashion_mnist
(train_x, train_y), (test_x, test_y) = mnist.load_data() #cargar los datos
train_x = train_x/255 ; test_y = test_y / 255 #normalizar

model = tf.keras.Sequential([ #creamos el modelo
                             tf.keras.layers.Flatten(),
                             tf.keras.layers.Dense(512, activation= "relu"),
                             tf.keras.layers.Dense(10,activation="softmax")
])
model.compile(optimizer="adam", loss = "sparse_categorical_crossentropy")
model.fit(train_x,train_y, epochs=10, callbacks = [callbacks])
```

## Hand writer digit recognizer

steps:
1. load the data set
2. divide in train an test
3. normalize de images
4. create the model, with layers and activations
5. compile de modelo with optimization and loss function
6. fit the model with epochs and call back
7. evaluate the model with test set
8. predict

```python
def train_mnist():
    class myCallback(tf.keras.callbacks.Callback):    
        def on_epoch_end(self, epoch, logs = {}):
            if(logs.get("accuracy") >= 0.99):
              print("\nReached 99% accuracy so cancelling training!")
              self.model.stop_training = True #detiene el training    

    mnist = tf.keras.datasets.mnist
    (x_train, y_train),(x_test, y_test) = mnist.load_data()

    x_train = x_train/255    
    
    callbacks = myCallback()            

    model = tf.keras.models.Sequential([                
        k.layers.Flatten(),
        k.layers.Dense(512, activation= "relu"),      
        k.layers.Dense(10, activation= "softmax")        
    ])

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    
    # model fitting
    history = model.fit(x_train,y_train,epochs = 10, callbacks = [callbacks])
    # model fitting
    return history.epoch, history.history['accuracy'][-1]
```

```python
train_mnist()
```
